##  *Image Brightening using Linear Transformation*

### **üìù Problem Description:**

You are given a color image represented in RGB format. Your task is to apply a **linear brightness enhancement** by adding a constant value $c$ to each pixel of the image using the transformation:

$$
g(x, y) = f(x, y) + c
$$

Where:

* $f(x, y)$: original pixel value at position $(x, y)$
* $c$: a positive constant to increase brightness
* $g(x, y)$: resulting pixel value after transformation

This operation must be applied to **all three channels** (Red, Green, and Blue). Ensure the output image is clipped to stay within the valid pixel range $[0, 255]$.

Finally, display both the original and the transformed image side by side.

---

### ‚úÖ Requirements:

1. Load a color image using OpenCV.
2. Convert the image from BGR to RGB format.
3. Apply the transformation $g(x, y) = f(x, y) + c$ to every pixel.
4. Clip all values to stay within 0‚Äì255.
5. Display both the original and modified images using `matplotlib`.

---

### üì• Input:

* A color image file (e.g., `image.jpg`)
* A scalar brightness value `c` (e.g., `c = 50`)

---

### üì§ Output:

* The brightened image after applying the transformation.
* A side-by-side comparison of the original and transformed images.

---

### üí° Hint:

Use `np.clip()` and `astype(np.uint8)` to ensure the final image is displayable.

---

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load color image (BGR format)
img_bgr = cv2.imread("Images/Flower1.jpg")

# Convert BGR to RGB
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Define constant c to add
c = 100

# Apply g(x,y) = f(x,y) + c (broadcasted over all channels)
#g_rgb = np.clip(img_rgb + c, 0, 255).astype(np.uint8)
g_rgb =  img_rgb+c

red_img=img_rgb[:, :, 0]
green_img=img_rgb[:, :, 1]
blue_img=img_rgb[:, :, 2]

# Show original and brightened image
plt.figure(figsize=(10, 5))

plt.subplot(2, 3, 1)
plt.title("Original RGB")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(2, 3, 2)
plt.title(f"g(x,y) = f(x,y) + {c}")
plt.imshow(g_rgb)

plt.subplot(2, 3, 3)
plt.title(f"Red Channel")
plt.imshow(red_img, cmap='Reds')

plt.subplot(2, 3, 4)
plt.title(f"Green Channel")
plt.imshow(green_img, cmap='Greens')

plt.subplot(2, 3, 5)
plt.title(f"Blue Channel")
plt.imshow(blue_img, cmap='Blues')

plt.tight_layout()
plt.show()
```

## Add Two Color Images 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load two color images (must be the same size)
img1 = cv2.imread("Images/RGB_Image.jpg")
img2 = cv2.imread("Images/black.png")

# Resize second image to match first (optional safeguard)
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

# Convert both from BGR to RGB for matplotlib display
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# Add the two image matrices element-wise (with clipping)
img_added = img1_rgb+img2_rgb

# Display images
plt.figure(figsize=(12, 5))

plt.subplot(1, 3, 1)
plt.title("Image 1")
plt.imshow(img1_rgb)
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Image 2")
plt.imshow(img2_rgb)
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Added Image (f1 + f2)")
plt.imshow(img_added)
plt.axis('off')

plt.tight_layout()
plt.show()
```
