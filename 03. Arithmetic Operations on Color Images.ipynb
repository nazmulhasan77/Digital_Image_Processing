
## **Element-wise Arithmetic Operations on Color Images**

---

### ðŸ“„ **Description:**

This problem involves performing basic arithmetic operationsâ€”**addition**, **subtraction**, **multiplication**, and **division**â€”on two color images using Python and OpenCV. The objective is to understand how pixel-wise arithmetic affects image appearance and to explore which operations exhibit **linearity** in the context of digital image processing.

Two images of the same size are loaded, converted from BGR to RGB format for visualization, and then processed using the following element-wise operations:

* **Addition**: Enhances brightness or fuses images.
* **Subtraction**: Highlights differences, useful for motion detection or edge extraction.
* **Multiplication**: Amplifies or masks regions; nonlinear in nature.
* **Division**: Normalizes or adjusts contrast; also nonlinear.

The resulting images are displayed side by side using Matplotlib for visual comparison. This exercise also investigates the **linearity** of these operations, identifying that only addition and subtraction maintain linearity, while multiplication and division do not.

---


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load two color images (must be the same size)
img1 = cv2.imread("Images/Flower1.jpg")
img2 = cv2.imread("Images/Flower.jpeg")

# Resize second image to match first image size (if needed)
#img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))
img2=cv2.resize(img2,(img1.shape[1],img1.shape[0]))

# Convert BGR (OpenCV default) to RGB (for matplotlib)
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)


#img_add=img1_rgb+img1_rgb
# Perform arithmetic operations using OpenCV (to handle clipping)
img_add = cv2.add(img1_rgb, img2_rgb)               # Addition (clipped at 255)
img_subtract = cv2.subtract(img1_rgb, img2_rgb)     # Subtraction (clipped at 0)
img_multiply = cv2.multiply(img1_rgb, img2_rgb)     # Multiplication (scaled automatically)
img_divide = cv2.divide(img1_rgb, img2_rgb + 1)     # Division (add 1 to avoid division by zero)

# Display results using matplotlib
titles = ['Original Image 1', 'Original Image 2', 'Addition', 'Subtraction', 'Multiplication', 'Division']
images = [img1_rgb, img2_rgb, img_add, img_subtract, img_multiply, img_divide]

plt.figure(figsize=(18, 8))
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
