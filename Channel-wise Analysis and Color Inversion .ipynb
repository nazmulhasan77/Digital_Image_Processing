
## **Problem Description:**

Write a Python program to load an RGB image, extract its Red, Green, and Blue channels, and generate its negative image. Display the original image, individual RGB channels using respective color maps, and the negative image using `matplotlib`.


## ðŸ“š Basic Theory

### ðŸ”¹ Digital Image Representation

A **digital image** is a two-dimensional array of pixels. Each pixel contains color information stored as numerical values. In **RGB images**, each pixel has three components:

* **Red**
* **Green**
* **Blue**

These components are typically stored in a 3D NumPy array of shape `(height, width, 3)`.

---

### ðŸ”¹ RGBA Images

Some images include an **alpha channel** for transparency, making the shape `(height, width, 4)`. The extra channel is removed when working only with RGB data.

---

### ðŸ”¹ Image Negative

The **negative of an image** is created by subtracting each pixelâ€™s value from the maximum possible value in the image. This inverts the color:

$$
\text{Negative}(x, y) = \text{MaxValue} - \text{Original}(x, y)
$$

For example:

* White becomes black
* Red becomes cyan
* Light areas become dark, and vice versa

---

### ðŸ”¹ Channel Separation

To better understand how colors are formed, each of the RGB channels can be separated and visualized:

* **Red channel** shows intensity of red pixels
* **Green channel** shows green intensities
* **Blue channel** shows blue intensities

This is useful for color analysis, feature extraction, and preprocessing in image processing tasks.

---

```python
import matplotlib.pyplot as plt

def main():
    # Load an image
    #img_path = '/Users/nazmulhasan77/Desktop/Code/Digital_Image_Processing/Images/RGB_Image.png'
    img_path = '/home/nazmul77/Desktop/Digital_Image_Processing/Images/RGB_Image.png'
    img_4D = plt.imread(img_path)
    print("Original image shape (with alpha if present):", img_4D.shape)
    
    # Convert 4D image into a 3D RGB image (remove alpha channel if present)
    img_3D = img_4D[:, :, :3]
    print("Converted RGB image shape:", img_3D.shape)
    
    # Print some pixel values
    print("Top-left 5x5 pixels (Red & Green channels):\n", img_3D[:5, :5, :2])
    print("Max value:", img_3D.max(), "| Min value:", img_3D.min())
    
    # Create the negative image
    negative_img = img_3D.max() - img_3D
    
    # Display original, channels, and negative
    plt.figure(figsize=(20, 8))

    plt.subplot(2, 3, 1)
    plt.imshow(img_3D)
    plt.title("Original RGB Image")

    plt.subplot(2, 3, 2)
    plt.imshow(img_3D[:, :, 0], cmap='Reds')
    plt.title("Red Channel")

    plt.subplot(2, 3, 3)
    plt.imshow(img_3D[:, :, 1], cmap='Greens')
    plt.title("Green Channel")

    plt.subplot(2, 3, 4)
    plt.imshow(img_3D[:, :, 2], cmap='Blues')
    plt.title("Blue Channel")

    plt.subplot(2, 3, 5)
    plt.imshow(negative_img)
    plt.title("Negative Image")

    plt.tight_layout()
    plt.show()
    plt.close()

if __name__ == '__main__':
    main()
```
