
## **Image Masking: Top-Left White Box Using Loops**

---

## Problem Description

You are given a color image represented as a matrix of pixels in RGB format. Your task is to apply a white rectangular mask to the **top-left corner** of the image by setting the pixel values in that region to white $[255, 255, 255]$.

Instead of using direct array slicing, you **must use nested loops** to iterate through each pixel in the defined rectangular region and set its color value manually.

---

## Input

* A color image (any size) loaded into your program.
* Two integers specifying the dimensions of the rectangular mask:

  * `box_width`: the width (in pixels) of the white box starting from the left edge.
  * `box_height`: the height (in pixels) of the white box starting from the top edge.

---

## Output

* The original image with a white rectangular box applied to the **top-left corner**, where the box size is `box_width Ã— box_height`.
* Display or save the resulting image.

---



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image (color)
img = cv2.imread("Images/black.png")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Copy for modification
masked_img = img_rgb.copy()

# Set size of the top-left box
box_height = 150
box_width = 150

# Apply white color using nested loops
for y in range(box_height):
    for x in range(box_width):
        masked_img[y, x] = [255, 255, 255]  # white in RGB

# Show result
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Top-Left White Box (Loop)")
plt.imshow(masked_img)
plt.axis('off')

plt.tight_layout()
plt.show()
```
