## **Pixel Value Histogram Analysis of an RGB Image**


### ðŸ“Œ **Problem Description:**

In digital image processing, a **histogram** represents the **frequency distribution of pixel intensity values** in an image. For an 8-bit image, each pixel value can range from 0 to 255. By counting how often each pixel value occurs, we can understand the image's brightness, contrast, and color distribution.

### ðŸ“š **Histogram Theory:**

* A **histogram** is a bar graph where:

  * **X-axis** represents pixel intensity values (0â€“255),
  * **Y-axis** shows the number of pixels with that intensity.
* For an RGB image:

  * Each channel (R, G, B) has its own histogram.
* Histogram analysis helps understand:

  * **Brightness**: More high values â†’ brighter image.
  * **Contrast**: Wide range of values â†’ high contrast.
  * **Color balance**: Differences between R, G, B distributions.

---

### ðŸ§  **Example:**

If an image contains:

```text
Red pixel values: [0, 0, 128, 255]
```

Then the red histogram would look like:

```python
pixel_count_r[0] = 2
pixel_count_r[128] = 1
pixel_count_r[255] = 1
```

````python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def main():

    img_path = "Images/Flower.jpeg"
    img = cv2.imread(img_path)
    # Convert BGR to RGB for correct channel labeling
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    height, width, _= img_rgb.shape
    x=np.arange(0,256,1)

    pixel_count_r= [0]*256
    pixel_count_g= [0]*256
    pixel_count_b= [0]*256

    height=img_rgb.shape[0]
    width=img_rgb.shape[1]

    # Count each channel's pixel value
    for i in range(height):
        for j in range(width):
            pixel_count_r[img_rgb[i,j,0]]+=1
            pixel_count_g[img_rgb[i,j,1]]+=1
            pixel_count_b[img_rgb[i,j,2]]+=1


    # for i in range(height):
    #     for j in range(width):
    #         r = img_rgb[i, j, 0]
    #         g = img_rgb[i, j, 1]
    #         b = img_rgb[i, j, 2]
    #         pixel_count_r[r] += 1
    #         pixel_count_g[g] += 1
    #         pixel_count_b[b] += 1

    print(pixel_count_r)
    total_pixels_r = sum(pixel_count_r)
    print("Total Pixel Per Chanel: ",total_pixels_r)

    plt.figure(figsize=(40, 40))
    plt.subplot(3,1,1)
    plt.stem(x,pixel_count_r,label="Red")
    plt.legend()

    plt.subplot(3,1,2)
    plt.plot(pixel_count_g, color='green', label='Green')
    plt.legend()

    plt.subplot(3,1,3)
    plt.plot(pixel_count_b, color='blue', label='Blue')

    plt.legend()
    plt.show()

if __name__ == '__main__':
    main()
````
